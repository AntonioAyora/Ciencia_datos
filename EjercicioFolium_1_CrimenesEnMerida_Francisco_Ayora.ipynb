Con los datos del archivo "asaltos_yucatan.csv", mostrar en un mapa la ubicación de los asaltos de los delitos en Mérida, de ser posible muestre como _popup_ la categoría.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar el archivo datos/asaltos_yucatan.csv en un DataFrame llamado data
# si es necesario usar la opcion encoding = "ISO-8859-1"
data = pd.read_csv('datos/asaltos_yucatan.csv', encoding = "ISO-8859-1")
```

```python
#Mostrar las primeras filas del DataFrame data
data.head()
```

```python
df = data[['Categoria','lat','long']]
df
```

```python
loc = df[ ['lat', 'long'] ].to_numpy()
loc
```

```python
import folium
```

```python
map_yuc = folium.Map(location=[20.9785, -89.6332],zoom_start=9)


for i in range(loc.shape[0]):
    folium.Marker(
        location = loc[i].tolist(),
        popup = df['Categoria'].values[i],
        icon=folium.Icon(color='blue',icon='none'),
    ).add_to(map_yuc)

map_yuc
```
