## Ejercicio 1:  Datos de la estación meteorologica de Progreso, Yucatán

Fuente:  https://smn.conagua.gob.mx/es/observando-el-tiempo/estaciones-meteorologicas-automaticas-ema-s

```python
df= pd.read_csv("datos/Estacion_PROGRESO_90_dias_19Nov.csv")
df.head(5)
```

<div class="alert alert-info">

<b>EJERCICIO</b>:

 <ul>
  <li> Configurar la columna de 'FechaLocal' como índice temporal  </li>
</ul>
</div>

```python
df['FechaLocal'] = pd.to_datetime(df['FechaLocal'], format= '%d/%m/%Y %H:%M')
```

```python
df = df.set_index('FechaLocal')
```

```python
df.head()
```

<div class="alert alert-success">
    
a). Graficar la Precipitación total por día, semana y mes (graficas separadas)
    
</div> 

```python
month = df['Precipitacion(mm)'].resample('M').sum()
month.plot(style='g--',figsize=(10,5))
plt.xlabel('Meses')
plt.ylabel('Precipitacion (mm)')
plt.title("Precipitcion total por mes")
plt.show()
```

```python
semana = df['Precipitacion(mm)'].resample('W').sum()
semana.plot(style='g--',figsize=(10,5))
plt.xlabel('Meses')
plt.ylabel('Precipitacion (mm)')
plt.title("Precipitcion total por semana")
plt.show()
```

```python
dia = df['Precipitacion(mm)'].resample('D').sum()
dia.plot(style='r--',figsize=(10,5))
plt.xlabel('dia')
plt.ylabel('Precipitacion (mm)')
plt.title("Precipitcion total por dia")
plt.show()
```

<div class="alert alert-success">
    
b). En una misma grafica muestre la media, el valor mínimo y el máximo de la temperatura del aire
    
</div> 

```python
valores = df['TemperaturaAire(C)'].resample('M').agg(['mean', 'min','max'])

valores.plot(style=['y:','b--','r-'],figsize=(10,5))
plt.xlabel('valores')
plt.ylabel('Meses')
plt.title("Media, valores mínimo y máximo")
plt.show()
```

<div class="alert alert-success">
    
c). Encontrar el día que en suma recibió mayor radiación solar; con este día haga una grafica con el parámetro de la mediana para cada 30 minutos.
    
</div> 

```python
m= df['RadiacionSolar(W/m2)'].resample('D').sum()
m.reset_index().sort_values('RadiacionSolar(W/m2)',ascending=False)
```

```python
day= df['RadiacionSolar(W/m2)']['2020-08-22'].resample('30T').sum()

day.plot(style=['g--'],figsize=(10,5))
plt.xlabel('Valores')
plt.ylabel('Mediana')
plt.title("Mediana de cada Media hora")
plt.show()
```

# Ejercicio 2: Incendios en México

```python
df= pd.read_csv("datos/Serie_historica_anual_incendios_2017.csv")
df.head(5)
```

<div class="alert alert-info">

<b>EJERCICIO</b>:

 <ul>
  <li> Configurar la columna de 'Fecha Inicio' como índice temporal desde el inicio de su lectura </li>
</ul>
</div>

```python
#  'Fecha Inicio' tiene índice 13
df = pd.read_csv("datos/Serie_historica_anual_incendios_2017.csv", index_col=13, parse_dates=True)
df.head()
```

Para poder realizar la segmentación(slicing) de días, semanas o meses será necesario ordenar los índices por fecha.

```python
df = df.sort_index()
df.tail()
```

<div class="alert alert-success">
    
d). Graficar la superficie Total de hectáreas afectadas en todo el país por día
    
</div> 

```python
area = df['Total Hectareas'].resample('D').sum()

area.plot(style=['g-'],figsize=(10,5))
plt.xlabel('Area')
plt.ylabel('Meses')
plt.title("Area afectada por los incendios")
plt.show()
```

<div class="alert alert-success">
    
e). Graficar la superficie total de hectáreas afectadas sólo en Yucatán por mes
    
</div> 

```python
df_yucatan=df[df['Estado']=='Yucatán']['Total Hectareas'].resample('M').sum()
df_yucatan.plot(style=['g--'],figsize=(10,5))
plt.xlabel('Area')
plt.ylabel('Meses')
plt.title("Area afectada por los incendios en Yucatan")
plt.show()
```

<div class="alert alert-success">
    
f). Graficar la superficie total de hectáreas afectadas por semana sólo en Jalisco para los meses abril, mayo, junio y julio 
    
</div> 

```python
df_jalisco=df[df['Estado']=='Jalisco']
mon=df_jalisco[df_jalisco.index.month.isin([4,5,6,7])]['Total Hectareas'].resample('W').sum()

mon.plot(style=['y--'],figsize=(10,5))
plt.xlabel('Area Afectada')
plt.ylabel('Mese de Abril, Mayo, Junio y Julio')
plt.title("Area afectada por los incendios en Jalisto de los meses de Abril, Mayo, Junio y Julio")
plt.show()
```

# Ejercicio 3: Pacientes confirmados con Covid en México


Fuente: https://datos.covid-19.conacyt.mx/#DownZCSV

<div class="alert alert-info">

<b>EJERCICIO</b>:

 <ul>
  <li> Para el análisis por fechas   </li>
</ul>
</div>

Como veremos a continuacón los datos de fecha se encuentran en las columnas, por lo que será necesario invertir su posición

```python
df= pd.read_csv("datos/Casos_Diarios_Estado_Nacional_Confirmados_20201118.csv")
df.head()
```

Conservaremos solo el nombre de la entidad y las columnas de las fechas

```python
df=df.loc[:,'nombre':]
df.head()
```

Llevamos a cabo la operacion Transpuesta, es decir intercambiar filas por columnas

```python
df = df.set_index('nombre').T
df.head()
```

El resto es ajustar el nombre de nuestra columna que será el indice y configurarla con pd.to_datetime 

```python
df=df.reset_index()
df.head()
```

```python
df = df.rename(columns = {'index': 'Fecha'})
df.head()
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], format= '%d-%m-%Y')
```

```python
df = df.set_index('Fecha')
```

```python
df
```

<div class="alert alert-success">
    
a). Graficar el total de pacientes confirmados nivel Nacional por semana, desde el 15 de enero hasta el 14 de noviembre
    
</div> 

```python
eno = df['2010-01-14':'2020-11-14']['Nacional'].resample('W').sum()

eno.plot(style=['ro'],figsize=(10,5))
plt.xlabel('Infectados')
plt.ylabel('Meses')
plt.title("Pacientes confirmados nivel Nacional por semana, desde el 15 de enero hasta el 14 de noviembre")
plt.show()
```

<div class="alert alert-success">
    
b). Graficar el total de pacientes confirmados cada 2 días en Yucatan (en todos los datos)
    
</div> 

```python
yuc= df['YUCATAN'].resample('2D').sum()

yuc.plot(style=['r--'],figsize=(10,5))
plt.xlabel('Infectados')
plt.ylabel('Dias')
plt.title("Total de pacientes confirmados cada 2 días en Yucatan ")
plt.show()
```

<div class="alert alert-success">
    
c). Encontrar la semana con más pacientes confirmados en Yucatán
    
</div> 

```python
m=df['YUCATAN'].resample('W').sum()
m.reset_index().sort_values('YUCATAN',ascending=False).head()
```

<div class="alert alert-success">
    
d). Encontrar la semana con más pacientes confirmados a nivel Nacional
    
</div>

```python
m=df['Nacional'].resample('W').sum()
m.reset_index().sort_values('Nacional',ascending=False).head()
```
